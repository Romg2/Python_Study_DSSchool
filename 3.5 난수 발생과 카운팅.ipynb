# 데이터 샘플링

````python
# numpy.random.choice(a, size = None, replace = True, p = None)
# a: 배열 입력, 정수 입력시 arrange(a)로 인식
# size: 추출할 샘플 갯수
# replace: 복원 추출 여부
# p: 배열의 각 데이터가 추출 될 확률
````

````python
# shuffle과 같음
np.random.choice(5,5,replace=False)
````

````python
# 일부 추출
np.random.choice(5,2,replace=False)
````

````python
# 복원 추출
np.random.choice(5,7)
````

````python
# 확률을 부여해서 복원 추출
np.random.choice(5,10, p = [0.1, 0, 0.4, 0.5, 0])
````

# 난수 생성

````python
# rand: 0~1 사이 uniform
# randn: standard normal
# ranint: 0~정수 사이 uniform
````

````python
# 균일 분포
print(np.random.rand(5))
print(np.random.rand(2,2))
````

````python
# 표준 정규 분포
print(np.random.randn(5))
print(np.random.randn(2,2))
````

````python
# 균일 분포2
print(np.random.randint(10, size=5)) # 0 ~ 10 사이
print(np.random.randint(10, 20, size=5)) # 10 ~ 20 사이
print(np.random.randint(10, 20, size=(2, 2))) # 10 ~ 20 사이 행렬
````

# 연습 문제 3.5.2

```

가격이 10,000원인 주식이 있다. 이 주식의 일간 수익률(%)은 기댓값이 0%이고 표준편차가 1%인 표준 정규 분포를 따른다고 하자.
250일 동안의 주가를 무작위로 생성하라
```

````python
price = np.ones(250) * 10000
x = np.random.randn(250)

# 전일 종가대비로 작성하지 않고 다음과 같은 수식으로 계산
# 10,000원 - (10,000원 x 수익률)
price - (price * x / 100)
````

# 정수 데이터 카운팅

````python
# 정수 데이터 카운팅
np.unique([11, 11, 2, 2, 34, 34])
````

````python
# 데이터와 그에 따른 개수
x = np.array(['a', 'b', 'b', 'c', 'a'])
index, count = np.unique(x, return_counts=True)

print(index, count)
````

````python
# 주사위 나온 눈의 개수
# bincount에서 minlength 지정시 0 ~ minlength -1 만큼의 요소별 개수가 출력
# 실제로 데이터가 없어도 개수 0으로 출력
dice = np.random.randint(1,7, size = 5)
print(dice)
print(np.bincount(dice, minlength=10))
````
